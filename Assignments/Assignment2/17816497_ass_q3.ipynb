```python
# Pratham Shukla | 17816497
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x,y,alpha):
    if x <= 0:
        return y + alpha/(1-x)
    elif 0 < x < (alpha + y):
        return alpha + y
    elif x >= alpha + y:
        return -1
```

```python
mu = 0.001
alphas = np.linspace(8, -1, 10)
sigmas = np.linspace(-1, 1, 11)
n = 20000
fig, axs = plt.subplots(nrows = 10, ncols = 11, figsize = (20,20))
plt.subplots_adjust(hspace = 0.5)

for j in range(10):
    alpha = alphas[j]
    for k in range(11):
        sigma = sigmas[k]
        x = np.zeros(n)
        y = np.zeros(n)
        for i in range(n-1):
            x[i+1] = f(x[i], y[i], alpha)
            y[i+1] = y[i] - mu*(x[i] + 1) + mu*sigma

        s = '=' + str("{0:.2f}".format(alpha)) + ',=' + str("{0:.2f}".format(sigma))
        axs[j,k].plot(x[17000:])
        axs[j,k].set_yticks([-1,0,1,2])
        axs[j,k].set_xticks([])
        axs[j,k].set_title(s)
```

As visible from the plots over the alpha-sigma space, the behaviors and the parameter ranges (, ) are such that,
1.    + 5 < 4, for silence
2.   ( + 5 > 4)  ( - 5 < 5), for bursts of spikes
3.   ( - 5 > 5)  ( + 5 > 4), for spiking

It should be noted that the constants in the above equations are dependent on the resolution of the analysis. Better estimates can be obtained with finer resolutions.



```python
# Individual examples

# Silence at  = 1,  = -0.4
n = 10000
alpha = 1
sigma = -0.4
x = np.zeros(n)
y = np.zeros(n)
for i in range(n-1):
    x[i+1] = f(x[i], y[i], alpha)
    y[i+1] = y[i] - mu*(x[i] + 1) + mu*sigma
plt.plot(x[7000:])
plt.yticks([-2, -1, 0, 1, 2]) 
plt.ylabel('$x_n$')
plt.xlabel('n')
plt.title('Silence at  = ' + str(alpha) + ',  = ' + str(sigma))
plt.show()

# Bursts of spikes at  = 7,  = -0.2
n = 10000
alpha = 7
sigma = -0.2
x = np.zeros(n)
y = np.zeros(n)
for i in range(n-1):
    x[i+1] = f(x[i], y[i], alpha)
    y[i+1] = y[i] - mu*(x[i] + 1) + mu*sigma
plt.plot(x[7000:])
plt.yticks([-2, -1, 0, 1, 2])
plt.ylabel('$x_n$')
plt.xlabel('n')
plt.title('Bursts of Spikes at  = ' + str(alpha) + ',  = ' + str(sigma))
plt.show()

# Bursts of spikes at  = 6,  = 0.8
n = 10000
alpha = 6
sigma = 0.8
x = np.zeros(n)
y = np.zeros(n)
for i in range(n-1):
    x[i+1] = f(x[i], y[i], alpha)
    y[i+1] = y[i] - mu*(x[i] + 1) + mu*sigma
plt.plot(x[9000:])
plt.yticks([-2, -1, 0, 1, 2])
plt.ylabel('$x_n$')
plt.xlabel('n')
plt.title('Spiking at  = ' + str(alpha) + ',  = ' + str(sigma))
plt.show()
```
